```python
# 4. 
# Define a function that accepts a string that is a word. 
# The function should capitalize the first letter of the word 
# if the word starts with a consonant.

vowels = ["a", "e", "i", "o", "u"]

def start_with_consonant(string):
  for char in string.lower():
      if char[0].lower() not in vowels:
         return string.title()
      else:
        return False
  

start_with_consonant('Any string')
start_with_consonant('this is a string')
```

```python
## Alternative as instructed:

def capitalize_starting_consonant(string):
    if type(string) != str:
        return False
    first_letter = string[0]
    if is_consonant(first_letter):
        string = string.capitalize()
    return string

capitalize_starting_consonant("Is coffee")
```

```python
# 5. 
# Define a function named calculate_tip. 
# It should accept a tip percentage (a number between 0 and 1) 
# and the bill total, and return the amount to tip.

pre_tip_bill = float(input('Please enter your total bill.'))
tip_requested = float(input('Using a decimal between 0 and 1, what % would you like to tip?'))

def calculate_tip():
    if float(tip_requested) < 0 or float(tip_requested >1):
        return 'The tip percentage must be between 0 and 1'
    tip_amount = pre_tip_bill * tip_requested
    return tip_amount




def calculate_total_bill():
    if float(pre_tip_bill) != float:
        return 'The bill must be entered as a numeric value.'
    tip_amount = pre_tip_bill * tip_requested
    total_bill = pre_tip_bill + tip_amount
    return total_bill

print(f'The amount you should tip is $ {calculate_tip(): .2f}. \nThe total amount for your bill and tip is ${calculate_total_bill(): .2f}.')
```

```python
# 6.
# Define a function named apply_discount. 
# It should accept a original price, and a discount percentage, 
# and return the price after the discount is applied.


def apply_discount():
    original_price = float(input('Please enter the original price for the item.'))
    discount_percentage = float(input('Using a decimal between 0 and 1, what % is the discount?'))
    discount = original_price * discount_percentage
    new_price = original_price - discount
    return(f'The total after discount is ${new_price: .2f}.')

apply_discount()
```

```python
#10. 
# Define a function named normalize_name. 
#-It should accept a string and return a valid python identifier, 
#-anything that is not a valid python identifier should be removed
#-leading and trailing whitespace should be removed
#-everything should be lowercase
#-spaces should be replaced with underscores

# for example:
# Name will become name
# First Name will become first_name
# % Completed will become completed

import keyword

def check_beginning_and_end(string):
    while string[0].isalpha() == False:
        string = string[1:] 
    while string[-1].isalnum() == False:
        string = string[:-1]
    return string

def handle_special_characters(string):
    new_string = ''
    for char in string: 
        if char.isalnum() == True:
            new_string += char
        elif char in ' _':
            new_string += '_'    
    return new_string

def is_keyword(string):
    if string in keyword.kwlist: 
        return True
    else:
        return False

def normalize_name(string):
    string = check_beginning_and_end(string)
    string = handle_special_characters(string)
    if not is_keyword(string):
        return string.lower()
    else:
        return normalize_name(input())
```

```python
# 11. 
# Write a function named cumulative_sum that accepts a list of numbers 
# and returns a list that is the cumulative sum of the numbers in the list.
# cumulative_sum([1, 1, 1]) returns [1, 2, 3]
# cumulative_sum([1, 2, 3, 4]) returns [1, 3, 6, 10]

def cumulative_sum(lists):
    cu_list = []
    length = len(lists)
    cu_list = [sum(lists[0:x:1]) for x in range(0, length+1)]
    return cu_list[1:]

cumulative_sum([3,4,5])
```
